### groupby

```python
from itertools import groupby
```

```python
def smaller_than_3(x):
    return x<3
```

```python
a=[1,2,3,4]
```

```python
group_obj=groupby(a,key=smaller_than_3)
for key,value in group_obj:
    print(key,list(value))
```

```python
a=[1,2,3,4,5]
```

```python
group_obj=groupby(a,key=lambda x:x<3)
for key,value in group_obj:
    print(key,list(value))
```

```python
persons=[{'name':'manvi','age':22},{'name':'bhoomi','age':21},
         {'name':'megha','age':22},{'name':'rithu','age':4}]
```

```python
group_obj=groupby(persons,key=lambda x:x['age'])
for key,value in group_obj:
    print(key,list(value))
```

```python
names=[{'name':'bhavana','age':21},{'name':'suma','age':27},
      {'name':'bhoomika','age':22},{'name':'bhavana','age':67}]
```

```python
gp_obj=groupby(names,key=lambda x:x['name'])
for key,value in gp_obj:
    print(key,list(value))
```

### infinite iterators

```python
from itertools import count,cycle,repeat
```

```python
for i in count(10):
    print(i)
    if i==15:
        break
```

```python
a=[1,2,3]
```

```python
for i in repeat(1,5):
    print(i)
```

```python
for i in cycle(a):
    print(i)
```
